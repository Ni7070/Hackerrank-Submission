**Nested Lists**

Given the names and grades for each student in a class of  students, store them in a nested list and print the name(s) of any student(s) having the second lowest grade.

Note: If there are multiple students with the second lowest grade, order their names alphabetically and print each name on a new line.

```python
#records = [["chi", 20.0],["alpha", 50.0], ["beta", 50.0]]
marks = []

records = []
n = int(input())
records = [[input(), float(input())] for _ in range(n)]
# for _ in range(int(input())):
#     name = input()
#     score = float(input())
#     records.append([name, score])

print(records)

for i in range(len(records)):
  marks.append(records[i][1])
print(marks)
marks.sort()
setOfMarks=set(marks)

print(setOfMarks)
listOfMarks = list(setOfMarks)

listOfMarks.sort()
print(listOfMarks)
secondLow = listOfMarks[1]
print(secondLow)
l=[]
for j in range(len(records)):
  if records[j][1] == secondLow:
    l.append(records[j][0])
l.sort()
for x in l:
  print(x)
```
