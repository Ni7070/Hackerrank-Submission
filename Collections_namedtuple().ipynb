```python
from collections import namedtuple

n = int(input())
Student = namedtuple('Student', input().split()) #Student function, parameter will be taken as input
l = 0                                            #counter
for i in range(n):
  p,q,r,s = input().split()                      # 4 variables for marks, name, id, class
  xyz = Student(p,q,r,s)                         # after passing to the Student function, result is temporary saved in xyx
  l = l + int(xyz.MARKS)                         # adding those marks one at a time
                                 
avgMark = round(l/n, 2)                          # average is rounded to 2 place after decimal
print(avgMark)
```
