```python
lis = []
N = int(input())
for _ in range(N):
    comm = list(map(str,input().strip().split(" ")))  # strip() deletes the first and last space of a sentence
    if comm[0] == 'insert':           # split(" ") splits when gets any whitespace
        lis.insert(int(comm[1]), int(comm[2]))
    elif comm[0] == 'print':
        print(lis)
    elif comm[0] == 'remove':
        lis.remove(int(comm[1]))
    elif comm[0] == 'append':
        lis.append(int(comm[1]))
    elif comm[0] == 'sort':
        lis.sort()
    elif comm[0] == 'pop':
        lis.pop()
    elif comm[0] == 'reverse':
        lis.reverse()
```
