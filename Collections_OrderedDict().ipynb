```python
from collections import OrderedDict

n = int(input())
od = OrderedDict()
for i in range(n):
  item, _, price = input().rpartition(' ') # there can be more than 1 space,
                                           # we need to split through the last occured space, rpartition(' ') dit that

  if item in od.keys():                   # if the same key is repeated
    od[item]= od[item] + int(price)       # price is added to its first price
  else:
    od[item]= int(price)                  # or price(value) is assigned to the opposite of item(key)
    
for item, price in od.items():            # for the sake of matched output
    print(item, price)
```
